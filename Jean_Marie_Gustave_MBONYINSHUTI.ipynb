## Data exploration and visualization

```python
vw_df.plot(kind='line', x='Final_Price', y='Final_Trip_Time (Minutes)', figsize=(25,10))
plt.show()
```

```python
vw_df.plot(kind='line', x='Final_Price', y='Final_Trip_Distance (KM)', figsize=(25,10), color='black')

plt.show()
```

```python
vw_df.plot(kind='line', x='Date', y='Final_Trip_Distance (KM)', figsize=(25,10), color='green')

plt.show()
```

```python
vw_df.plot(kind='line', x='Date', y='Final_Trip_Time (Minutes)', figsize=(25,10), color='red')

plt.show()
```

```python
print(np.array(vw_df['Driver_Name'].dropna().unique()))
vw_df['Final_Trip_Distance (KM)'].groupby(vw_df['Driver_Name']).sum()
```

```python
print(np.array(vw_df['Driver_Name'].dropna().unique()))
vw_df['Final_Trip_Time (Minutes)'].groupby(vw_df['Driver_Name']).sum()
```

```python
df = pd.DataFrame(vw_df['Final_Trip_Distance (KM)'].groupby(vw_df['Driver_Name']).sum())
df.plot(kind = 'bar', figsize=(25,10))
plt.show()
```

```python
df = pd.DataFrame(vw_df['Final_Trip_Time (Minutes)'].groupby(vw_df['Driver_Name']).sum())
df.plot(kind = 'bar', figsize=(25,10), color='brown')
plt.show()
```
