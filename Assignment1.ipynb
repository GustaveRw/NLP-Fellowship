TODO 3 - Create a new json consisting of aggregated data of the products: {category, brand : list of products, average price}

```python
response = requests.get("https://dummyjson.com/products")
data = response.json()
products = []
all_products = [product for product in data['products']]
def get_category(all_items):
    all_category = []
    for all_item in all_items:
        all_category.append(all_item['category'])
    return all_category
mylist = list(dict.fromkeys(get_category(all_products)))
final_items = []
final_brand = []
final_price = []
def get_brand_same_category(category):
    for all_items in all_products:
        if all_items['category'] == category:
            final_brand.append(all_items['brand'])
    return final_brand
def get_brand_same_price(category):
    for all_items in all_products:
        if all_items['category'] == category:
            final_price.append(all_items['price'])
            final_brand.append(all_items['brand'])
    average_final = sum(final_price) / len(final_price)
    return average_final
for my in mylist:
    final_items.append({
        my: {
            'brand': list(dict.fromkeys(get_brand_same_category(my))),
            'average': get_brand_same_price(my)
        }
    })
```

```python
print(final_items)
```

TODO 4 - Write the new json to a file

```python
with open('/content/convert.txt', 'w') as convert_files:
    convert_files.write(json.dumps(final_items))
```

Organize the code into csv format in a way one can get information. Be creative with the presentation :) 3 drivers

```python
import urllib
import json
import pandas as pd
import requests
import csv


def get_drivers_data():
    response = requests.get('http://ergast.com/api/f1/2018/drivers.json')

    data = response.json()
    final_drivers = data['MRData']['DriverTable']['Drivers']
    return final_drivers

final_dict = []
for data in get_drivers_data():
    final_dict.append({
        'driver_name': data['driverId'],
        'nationality': data['nationality'],
        'code': data['code']
    })

def jsonExtracter(file_path):
    drivers_set = final_dict[:3]
    field_names = ['driver_name', 'nationality', 'code']
    with open(f'{file_path}', 'w') as csvfile:
        writer = csv.DictWriter(csvfile, fieldnames=field_names)
        writer.writeheader()
        writer.writerows(drivers_set)
        return drivers_set


jsonExtracter('/content/race.csv')
```
