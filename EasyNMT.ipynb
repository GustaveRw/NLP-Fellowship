```python
from easynmt import EasyNMT
```

```python
model = EasyNMT('opus-mt')
```

```python
sentences =  [
    'I love my life.', 'Conversation AI is part of NLP',
    'I met interesting people this month', 'I remember to brush my teeth in the morning',
    'Jesus is alive', 'My mother is my everything'
]
```

```python
translations_fr = model.translate(sentences, target_lang='fr')
translations_rw = model.translate(sentences, target_lang='rw')
translations_zh = model.translate(sentences, target_lang='zh')
```
