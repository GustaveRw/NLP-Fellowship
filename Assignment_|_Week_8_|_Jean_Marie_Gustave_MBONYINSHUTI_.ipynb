importing the necessary libraries

```python
!pip install transformers

from transformers import *

import tensorflow as tf
import tensorflow_datasets as tfds

tfds.disable_progress_bar()
```

Download and load the dataset

```python
!pip install datasets
from datasets import load_dataset

dataset = load_dataset("kinnews_kirnews", "kinnews_cleaned")
```

Set the pretrained model and tokenizer

```python

model_name = "roberta-base"
tokenizer = RobertaTokenizer.from_pretrained(model_name)
```

Preparing the data for training by encoding the text and labels and creating a Dataset object

```python
# Encode the text and labels
def encode(text, label):
    encoded_text = tokenizer.encode(text, max_length=512)
    return encoded_text, label


```

```python
# Create the Dataset object
dataset = dataset.map(encode)
```

training the model using the Dataset object and the RoBERTa-base model

```python
# Set the batch size and number of epochs
batch_size = 32
epochs = 1

# Create the model
model = TFRobertaForSequenceClassification.from_pretrained(model_name)

# Compile the model
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=3e-5),
    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    metrics=["accuracy"],
)

# Train the model
model.fit(dataset, batch_size=batch_size, epochs=epochs)
```
